```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
data.head(5)
```

```python
# adding the target column to the dataframe
data['label'] = breast_cancer_dataset.target
data.tail(5)
```

```python
# number of rows and cols in the dataset
data.shape
```

```python
# checking for missing values
data.isnull().sum()
```

```python
# statistical measures about the dataset
data.describe()
```

```python
data['label'].value_counts()
```

1->Bening

0->Malignant



```python
data.groupby('label').mean()
```

Seprating the features and target


```python
X= data.drop(columns='label', axis=1)
Y = data['label']
```

```python
print(X)
print(Y)
```

SPLITTING THE DATA INTO TRAINING AND TESTING DATA

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the Data


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

MODEL TRAINING


LOGISTIC REGRESSION

```python
model = LogisticRegression()
```

```python
# training the logisticRegression model using the training data

model.fit(X_train, Y_train)
```

MODEL EVALUATION

Accuracy Score

```python
# acccuracy of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(Y_train, X_train_prediction)
print(training_data_accuracy)
```

```python
# acccuracy of testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy= accuracy_score(Y_test, X_test_prediction)
print(testing_data_accuracy)
```

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up layers of neural network
# 1 input, 1 hidden, 1 output layer
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network

model.compile(
                optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy']
              )
```

```python
# training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

visualizing accuracy and loss


```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training_data', 'validation_data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['loss', 'val_loss'], loc='upper right')
```

**ACCURAACY OF MODEL ON TEST DATA**
